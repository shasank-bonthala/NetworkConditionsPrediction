```python
# f1
def max_size(x):
    nums = x.split(';')[:-1]
    nums_int = list(map(int, nums))
    return max(nums_int)
```

```python
plt.scatter(data['packet_loss_ratio'], data['packet_sizes'].apply(max_size))
plt.xlabel('Packet Loss Ratio\nFigure 1.1')
plt.ylabel('Maximum Packet Size (Bytes)')
plt.title('Max Packet Size vs. Packet Loss Ratio');
```

```python
# f2
def range_size(x):
    nums = x.split(';')[:-1]
    nums_int = list(map(int, nums))
    return max(nums_int) - min(nums_int)
```

```python
plt.scatter(data['packet_loss_ratio'], data['packet_sizes'].apply(range_size))
plt.xlabel('Packet Loss Ratio\nFigure 1.1')
plt.ylabel('Range in Packet Size (Bytes)')
plt.title('Range in Packet Size vs. Packet Loss Ratio');
```

```python
# Feature 3: Average of Packet Size for each interaction
def avg_size(x):
    nums = x.split(';')[:-1]
    numsInt = list(map(int, nums))
    return np.mean(numsInt)
```

```python
plt.scatter(data['packet_loss_ratio'], data['packet_sizes'].apply(avg_size))
plt.xlabel('Packet Loss Ratio\nFigure 1.2')
plt.ylabel('Mean Packet Size (Bytes)')
plt.title('Mean Packet Size vs. Packet Loss Ratio');
```

```python
# Feature 4: Longest Packet Duration
def packet_dur(x):
    nums = x.split(';')[:-1]
    numsInt = list(map(int, nums))
    return np.mean(np.diff(numsInt))
```

```python
plt.scatter(data['packet_loss_ratio'], data['packet_times'].apply(packet_dur))
plt.xlabel('Packet Loss Ratio\nFigure 1.3')
plt.ylabel('Mean Packet Duration (ms)')
plt.title('Mean Packet Duration vs. Packet Loss Ratio');
```

```python
# Feature 5: Total packet Direction
def total_packet_dir(x):
    dirs = x.split(';')[:-1]
    totalDirs = 0
    
    for i in dirs:
        if i == '1':
            totalDirs += 1
        elif i == '2':
            totalDirs -= 1
           
    return totalDirs
```

```python
plt.scatter(data['packet_loss_ratio'], data['packet_dirs'].apply(total_packet_dir))
plt.xlabel('Packet Loss Ratio\nFigure 1.4')
plt.ylabel('Packet Direction')
plt.title('Packet Direction vs. Packet Loss Ratio');
```

```python
# Feature 8: Interaction length
def interaction_length(x):
    times = x.split(';')[:-1]
    times2 = list(map(int, times))
    startTime = min(times2)
    endTime = max(times2)
           
    return endTime - startTime
```

```python
plt.scatter(data['packet_loss_ratio'], data['packet_times'].apply(interaction_length))
plt.xlabel('Packet Loss Ratio\nFigure 1.5')
plt.ylabel('Interaction Length (ms)')
plt.title('Interaction Length vs. Packet Loss Ratio');
```

```python
# f9
plt.scatter(data['packet_loss_ratio'], (data['1->2Pkts'] + data['2->1Pkts'])/data['packet_times'].apply(interaction_length))
plt.xlabel('Packet Loss Ratio\nFigure 1.6')
plt.ylabel('Packets Sent over Time (Bytes/ms)')
plt.title('Packets Sent over Time vs. Packet Loss Ratio');
```

```python
# f10
plt.scatter(data['packet_loss_ratio'], (data['1->2Bytes'] + data['2->1Bytes'])/data['packet_times'].apply(packet_dur))
plt.xlabel('Packet Loss Ratio\nFigure 1.2')
plt.ylabel('Mean Packet Size (Bytes)')
plt.title('Mean Packet Size vs. Packet Loss Ratio');
```
