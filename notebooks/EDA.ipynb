```python
data1 = pd.read_csv('DANEdata/20211015T175337_200-5000-iperf.csv')
data2 = pd.read_csv('DANEdata/20211023T012228_200-500-iperf.csv')
data3 = pd.read_csv('DANEdata/20211023T012231_200-750-iperf.csv')
data4 = pd.read_csv('DANEdata/20211104T232610_200-200-iperf.csv')
data5 = pd.read_csv('DANEdata/20211105T000043_200-1000-iperf.csv')
data6 = pd.read_csv('DANEdata/20211105T003333_200-2000-iperf.csv')
data7 = pd.read_csv('DANEdata/20211105T003334_200-1600-iperf.csv')
data8 = pd.read_csv('DANEdata/20211105T004748_200-1800-iperf.csv')
data9 = pd.read_csv('DANEdata/20211105T164434_200-3000-iperf.csv')
data10 = pd.read_csv('DANEdata/20211105T170902_200-4000-iperf.csv')
data11 = pd.read_csv('DANEdata/20211105T180745_200-7000-iperf.csv')
data12 = pd.read_csv('DANEdata/20211105T185733_200-8500-iperf.csv')
data13 = pd.read_csv('DANEdata/20211105T205042_200-9500-iperf.csv')
data14 = pd.read_csv('DANEdata/20211105T205041_200-10500-iperf.csv')
data15 = pd.read_csv('DANEdata/20211105T205041_200-11000-iperf.csv')
```

```python
#Adding the packet loss ratio for each dataset

packet_ratio = np.ones(310) * (1/5000)
data1['packet_loss_ratio'] = packet_ratio

packet_ratio2 = np.ones(314) * (1/500)
data2['packet_loss_ratio'] = packet_ratio2

packet_ratio3 = np.ones(314) * (1/750)
data3['packet_loss_ratio'] = packet_ratio3

packet_ratio4 = np.ones(317) * (1/200)
data4['packet_loss_ratio'] = packet_ratio4

packet_ratio5 = np.ones(317) * (1/1000)
data5['packet_loss_ratio'] = packet_ratio5

packet_ratio6 = np.ones(311) * (1/2000)
data6['packet_loss_ratio'] = packet_ratio6

packet_ratio7 = np.ones(311) * (1/1600)
data7['packet_loss_ratio'] = packet_ratio7

packet_ratio8 = np.ones(312) * (1/1800)
data8['packet_loss_ratio'] = packet_ratio8

packet_ratio9 = np.ones(310) * (1/3000)
data9['packet_loss_ratio'] = packet_ratio9

packet_ratio10 = np.ones(309) * (1/4000)
data10['packet_loss_ratio'] = packet_ratio10

packet_ratio11 = np.ones(309) * (1/7000)
data11['packet_loss_ratio'] = packet_ratio11

packet_ratio12 = np.ones(309) * (1/8500)
data12['packet_loss_ratio'] = packet_ratio12

packet_ratio13 = np.ones(306) * (1/9500)
data13['packet_loss_ratio'] = packet_ratio13

packet_ratio14 = np.ones(307) * (1/10500)
data14['packet_loss_ratio'] = packet_ratio14

packet_ratio15 = np.ones(308) * (1/11000)
data15['packet_loss_ratio'] = packet_ratio15
```

```python
data = pd.concat([data1,data2,data3,data4,data5,data6,data7,data8,data9,data10,data11,data12,data13,data14,data15])
data
```
